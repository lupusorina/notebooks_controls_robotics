```python
# Compiled with Python 3.8

import numpy as np
from systems import Dubins, Ackermann, AckermannVelDelay, TwoDimDoubleIntegratorNominal, TwoDimSingleIntegratorNominal
import matplotlib.pyplot as plt
```

```python
DT = 0.01 # integration step.
N = 1000 # nb of integration steps.
```

# Dubins Model

```python
system_dubins = Dubins()
state_list = np.zeros((N, system_dubins.xdim))
state = np.array([0, 0, 0])
action = np.array([1, 0.1])
for i in range(N):
    state, _, _ = system_dubins.dynamics(state, action, DT)
    state_list[i] = state

# Plot.
# We expect the system to follow a curve
plt.plot(state_list[:, 0], state_list[:, 1])
# Initial state.
plt.plot(state_list[0, 0], state_list[0, 1], 'ro', label='Initial state')
# Final state.
plt.plot(state_list[-1, 0], state_list[-1, 1], 'go', label='Final state')
plt.ylabel('y [m]')
plt.xlabel('x [m]')
plt.legend()
```
